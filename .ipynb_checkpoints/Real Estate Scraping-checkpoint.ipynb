```python
import requests
from bs4 import BeautifulSoup

page = requests.get('http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/')
content=page.content

soup=BeautifulSoup(content,'html.parser')

all=soup.find_all('div',{"class":'propertyRow'})

all[0].find('h4',{'class':'propPrice'}).text.replace('\n','').replace(' ','')
```

```python
prop_list =[]
# This is provided via waybackwhen and acts strange when trying to access it... Maybe that's why the same page appears all three times despite having different URLs.
base_url='http://www.pythonhow.com/real-estate/rock-springs-wy/LCWYROCKSPRINGS/#t=0&s='

page_nr=soup.find_all('a',{'class':'Page'})[-1].text
print(page_nr)

for page in range(0,(int(page_nr)*10),10):
    print(base_url+str(page)+'.html')
    r=requests.get(base_url+str(page)+'.html')
    c=r.content
    soup=BeautifulSoup(c,'html.parser')
    all=soup.find_all('div',{'class':'propertyRow'})
#     Build dictionaries for each page with 10 properties per dict.
    for item in all:
        d={}
        d['Address']=item.find_all('span',{'class':'propAddressCollapse'})[0].text
        try:
            d['Locality']=item.find_all('span',{'class':'propAddressCollapse'})[1].text
        except:
            d['Locality']=None
        d['Price']=item.find('h4',{'class':'propPrice'}).text.replace('\n','').replace(' ','')    
        try:
            d['Beds']=item.find('span',{'class':'infoBed'}).find('b').text
        except:
            d['Beds']=None

        try:
            d['Area']=(item.find('span',{'class':'infoSqFt'}).find('b').text)
        except:
            d['Area']=(None)

        try:
            d['Full Baths']=(item.find('span',{'class':'infoValueFullBath'}).find('b').text)
        except:
            d['Full Baths']=(None)

        try:
            d['Half Baths']=(item.find('span',{'class':'infoValueHalfBath'}).find('b').text)
        except:
            d['Half Baths']=(None)

        for col_group in item.find_all('div',{'class':'columnGroup'}):
            for feature_group, feature_name in zip(col_group.find_all('span',{'class':'featureGroup'}), col_group.find_all('span',{'class':'featureName'})):
                if "Lot Size" in feature_group.text:
                    d['Lot Size']=(feature_name.text)
        prop_list.append(d)
```

```python
import pandas
df=pandas.DataFrame(prop_list)
```
